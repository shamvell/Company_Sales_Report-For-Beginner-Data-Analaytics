```python
df.groupby("product_1")["product_2"].sum().sort_values(ascending=True)
#📝 Step-by-Step Description 
#📄 df:
#Refers to your Pandas DataFrame containing the data.

#🧩 .groupby("product_1"):
#Groups the data based on unique values in the column product_1.
#Each group contains all rows that have the same product_1 value.

#🎯 ["product_2"]:
#Selects the column product_2 on which you want to perform aggregation (in this case, sum).

#➕ .sum():
#Calculates the total (sum) of product_2 values for each group created from product_1.

#⬇️ .sort_values(ascending=True):
#Sorts the final result (summed values) in ascending order (smallest to largest).

#✅ Final Output:
#A Pandas Series where:

#The index is the unique values from product_1.

#The values are the total sum of corresponding product_2 entries.

#Sorted from lowest to highest sum.
 

```

```python
df.groupby("product_1").sum(numeric_only=True)
#📝 Step-by-Step Description 
#📄 df:
#Refers to your Pandas DataFrame.

#🧩 .groupby("product_1"):
#Groups the DataFrame based on unique values in the product_1 column.
#Each group will contain rows with the same value in product_1.

#➕ .sum(numeric_only=True):

#Computes the sum of all numeric columns for each group.

#Non-numeric columns (like strings) are ignored.

#numeric_only=True ensures that only integer or float columns are included in the sum.

#✅ Final Output:
#A DataFrame where:

#The index is the unique values from product_1.

#Each numeric column contains the total sum for that group.

#Non-numeric columns will not appear in the result.

```

```python
 #Step 1: Total Sales for Each Product
df.sum(numeric_only=True)
```

```python
#Step 2: Sort Products by Total Sales
df.sum(numeric_only=True).sort_values(ascending=False)
```

```python
#Step 3: Visualize Top 10 Products
import matplotlib.pyplot as plt

df.sum(numeric_only=True).sort_values(ascending=False).head(10).plot(kind="bar", title="Top 10 Best-Selling Products")
plt.ylabel("Total Sales")
plt.show()
```

```python

# 1. Bar Chart – Total Sales by Product (Most Common)python

import matplotlib.pyplot as plt

total_sales = df.sum(numeric_only=True).sort_values(ascending=False)

plt.figure(figsize=(14,6))
total_sales.plot(kind='bar')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
#✅ 2. Pie Chart – Share of Sales by Product

import matplotlib.pyplot as plt

# Get total sales by product
total_sales = df.sum(numeric_only=True)

# Plot pie chart
plt.figure(figsize=(10,8))
total_sales.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title("Sales Share by Product")
plt.ylabel('')  # Hides the y-label
plt.tight_layout()
plt.show()
```

```python
#✅ 3. Horizontal Bar Chart – Top 10 Products by Sales
import matplotlib.pyplot as plt

# Top 10 products
top_10_sales = df.sum(numeric_only=True).sort_values(ascending=False).head(10)

# Plot horizontal bar chart
plt.figure(figsize=(12,6))
top_10_sales.plot(kind='barh', color='skyblue')
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product")
plt.gca().invert_yaxis()  # Highest at top
plt.tight_layout()
plt.show()
```

```python
✅ 4. Heatmap – Correlation Between Products
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm", annot=False)
plt.title("Correlation Between Product Sales")
plt.tight_layout()
plt.show()


```

```python
✅ 5. Box Plot – Distribution of Sales per Product
import seaborn as sns
import matplotlib.pyplot as plt

# Convert wide to long format
df_long = df.melt(var_name='Product', value_name='Sales')

# Plot boxplot
plt.figure(figsize=(14,6))
sns.boxplot(data=df_long, x='Product', y='Sales')
plt.xticks(rotation=90)
plt.title("Sales Distribution per Product")
plt.tight_layout()
plt.show()
```

```python
✅ 6. Histogram – Frequency of Sales Values
plt.figure(figsize=(10,6))
df.sum(axis=1).plot(kind='hist', bins=30, color='purple')
plt.title("Distribution of Total Sales per Record")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
✅ 7. Pair Plot – Compare Top Products
import seaborn as sns

# Select top 4 products for comparison
top_products = df.sum(numeric_only=True).sort_values(ascending=False).head(4).index

# Plot pairplot
sns.pairplot(df[top_products])
plt.suptitle("Pairwise Comparison of Top Products", y=1.02)
plt.show()
```
